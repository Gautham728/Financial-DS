```python
import pandas as pd
iris = pd.read_csv('iris - iris.csv')
print(iris.head())
```

```python
#get columns to plot
columns = iris.columns.drop(['variety'])
#create x data
x_data = range(0,iris.shape[0])
#create figure and axis
fig, ax = plt.subplots()
#plot each column
for column in columns:
    ax.plot(x_data,iris[column],label=column)
#set title and legend
ax.set_title('iris - iris.csv')
ax.legend()
```

```python
#Bar Chart
wine_reviews = pd.read_csv('winemag-data-130k-v2.csv')
wine_reviews.head()
```

```python
#create a figure and axis
fig, ax = plt.subplots()
#count the occurance of each class
data = wine_reviews['points'].value_counts()
#get x and y data
points = data.index
frequency = data.values
#create a bar chart 
ax.bar(points, frequency)
#set title and labels
ax.set_title('Wine Reviews Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
import numpy as np
import matplotlib.pyplot as plt
objects = ('Python','Java','C++','Perl','Scala','Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
#Bar chart
#X Axis positions as first parameter list,it can be floating points
#Y values as 2 nd parameter list
#Alpha is transparency
#Alpha can be center or edge
#Color can be single value or list of color codes, for each
plt.bar(y_pos,performance,width=0.5,align='center',alpha=0.5,color=('r','black','g','yellow','b','orange'))
#To define labels for x axis values
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.xlabel('Programming languages')
plt.title('programming languages')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
countries = ['USA','Brazil','Spain','India','UK']
totalDeaths = [1234,3566,3789,1222,3876]
plt.bar(countries, totalDeaths, width=0.5, align='center',alpha=0.5,color=('r','g','b','orange','yellow'))
i=1.0
j=100
for i in range(len(countries)):
    plt.annotate(totalDeaths[i],(-0.1 + i,totalDeaths[i] + j))
plt.legend(labels=['totaDeaths'])
plt.title("Bar plot represnting the total deaths by top 7 contries")
plt.xticks(countries)
plt.xlabel('Countries')
plt.ylabel('TotalDeaths')
plt.savefig('1BarPlot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
countries = ['USA','Brazil','Spain','India','UK']
totalDeaths = [24523,34556,37892,34122,37887]
totalCases = [423784,466789,489789,441822,452876]
for i in range(len(countries)):
    plt.bar(countries[i],totalDeaths[i], bottom=totalCases[i]- totalDeaths[i],color = 'black')
    plt.bar(countries[i],totalCases[i]- totalDeaths[i],color = 'r')
plt.legend(labels = ['Total Deaths','Total Cases'])
plt.xlabel('Countries')
plt.ylabel('TotalCases')
plt.title("Bar plot represnting the total deaths by top 12 contries")
plt.savefig('3BarPlot.png')
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[15,10])
totalDeaths = [24523,34556,37892,34122,37887]
totalCases = [423784,466789,489789,441822,452876]
totalRecovery = [145523,245556,278592,245122,279586]
countries = ['USA','Brazil','Spain','India','UK']
X = np.arange(len(totalDeaths))
    
plt.bar(X, totalDeaths, color = 'black',alpha = 0.5, width = 0.25)
plt.bar(X +0.25, totalRecovery, color = 'g',alpha = 0.5, width = 0.25)
plt.bar(X +0.5, totalCases, color = 'b',alpha = 0.5, width = 0.25)

plt.legend(['TotalDeaths','TotalRecovery','TotalCases'])
plt.xticks([i +0.25 for i in range(5)],countries)
plt.title("Bar plot representing thw total deaths, total recovery and total cases of top 6 countries")
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('4BarPlot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#if using Juptier notebook included:
%matplotlib inline
#pie chart where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,50,45,10]

fig,ax = plt.subplots()
ax.pie(sizes,labels=labels, autopct='%1.2f%%')
ax.axis('equal')#Equal ascpects ensures the pie chart is equal
ax.set_title('Engineering Diciplines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#if using Juptier notebook included:
%matplotlib inline
#pie chart where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,50,45,10]
#exploded out the 'Chemical' pie piece by offseting it a greater amount
explode = (0.1,0.1,0.1,0.4)

fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels, autopct='%1.2f%%',startangle=90)
ax.axis('equal')#Equal ascpects ensures the pie chart is equal
ax.set_title('Engineering Diciplines')
plt.show()
```
